### Estado del arte:

#### Opción de reconocimiento de Audio de IOS14:
Esta opción fue introducida a los dispositivos iphone en la actualización IOS 14 en septiembre de 2020, la cual puede notificar al usuario de la detección de sonidos específicos como:
- Llantos de bebe
- Sonidos de animales
- Llave de agua
- Vehículos
- etc.

Para poder usar esta opción es necesario tener dos dispositivos con IOS (Iphone y Apple Watch) y dejar uno de estos en el lugar donde se desea utilizar el reconocimiento de sonido.
Esta opción tiene la desventaja de deshabilitar el uso del asistente virtual Siri.

#### Merlin Bird ID:
Creado por Cornell Lab de la Universidad Cornell en Nueva York, EE.UU., Es una aplicación para la identificación de distintas aves, lo cual lo puede hacer de 3 formas diferentes:
- Respondiendo una serie de preguntas
- Subiendo una fotografía a la aplicación
- Subiendo un archivo de audio con el canto del ave
Además tiene la capacidad de mostrar las aves encontradas cerca del usuario.
